```python
from scrappy import Scrapper
import pandas as pd
```

```python
url = 'https://fbref.com/en/comps/Big5/stats/players/Big-5-European-Leagues-Stats'

headers = {
        'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"}
```

```python
jogadores = Scrapper(link= url, headers= headers)

def makeDict(self, num_inicial, num_final):
    columns = self.makeCols()
    rows = self.makeContent(num_inicial, num_final) 
    dicionario = {}
    n = len(columns)

    splited = [rows[i::n] for i in range(n)]

    for i in range(0, 35):
        dicionario[columns[i]] = splited[i]

    return dicionario
```

```python
dicionario = makeDict(self= jogadores, num_inicial= 0, num_final= 2769)
df = pd.DataFrame(dicionario)
df.head()
```
