```python
import pandas as pd
from scrappy import Scrapper
```

# Status

```python
url = 'https://fbref.com/en/comps/Big5/2022-2023/stats/players/2022-2023-Big-5-European-Leagues-Stats'

headers = {
        'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"}
```

```python
players = Scrapper(url, headers)
dicionario = players.createDict(0,3004)
df = pd.DataFrame(dicionario)
```

```python
df[df['Squad'] == "Real Madrid"].head(26)
```

```python
df.drop(['xG', 'npxG', 'xAG', 'npxG+xAG', 'PrgC', 'PrgP', 'PrgR',
       'G+A-PK', 'xG+xAG'], inplace= True, axis= 1)
df.to_csv('Desktop\\scrapper\\soccer_extract\\Extract_Fotbal_Players_data\\output\\2022-2023_footbal_players_status.csv')
```

# Defensive

```python
def_url = 'https://fbref.com/en/comps/Big5/2022-2023/defense/players/2022-2023-Big-5-European-Leagues-Stats'
players = Scrapper(def_url, headers)
dicionario = players.createDict(0,3004)
def_df = pd.DataFrame(dicionario)
```

```python
print(def_df.columns)
```

```python
def_df.drop(['Tkl+Int'], inplace= True, axis= 1)
rename_columns = {
    'Def 3rd': 'Tkl_Def',
    'Mid 3rd': 'Tkl_Mid',
    'Att 3rd': 'Tkl_Att'
}
def_df.rename(columns= rename_columns, inplace= True)
```

```python
def_df[def_df['Squad'] == 'Roma'].head(30)
```

```python
def_df.to_csv('Desktop\\scrapper\\soccer_extract\\Extract_Fotbal_Players_data\\output\\2022-2023_footbal_players_defensive_status.csv')
```

# Pass

```python
pass_url = 'https://fbref.com/en/comps/Big5/2022-2023/passing_types/players/2022-2023-Big-5-European-Leagues-Stats'
players = Scrapper(pass_url, headers)
dicionario = players.createDict(0,3004)
pass_df = pd.DataFrame(dicionario)
```

```python
print(pass_df.columns)
```

```python
rename_columns = {
    'Live': 'Live_Pass',
    'Dead': 'Dead_Pass',
    'Att' : 'Pass_Att',
    'Blocks' : 'Blocked_Pass',
    'Off' : 'Pass_off'
}
pass_df.rename(columns= rename_columns, inplace= True)
```

```python
pass_df[pass_df['Squad'] == 'Manchester City'].head(24)
```

```python
pass_df.to_csv('Desktop\\scrapper\\soccer_extract\\Extract_Fotbal_Players_data\\output\\2022-2023_footbal_players_pass.csv')
```

# Join

```python
df_join = pd.merge(df, def_df[['Player','Tkl',
       'TklW', 'Tkl_Def', 'Tkl_Mid', 'Tkl_Att', 'Att', 'Tkl%', 'Lost',
       'Blocks', 'Sh', 'Pass', 'Int', 'Clr', 'Err']], on= 'Player', how='left')
```

```python
df_join = pd.merge(df_join, pass_df[['Player', 
                                     'Live_Pass', 'Dead_Pass', 'Pass_Att',
                                     'Cmp', 'Pass_off', 'Blocked_Pass']], on= 'Player', how='left')
```

```python
keep_list = ["Player", "Squad"]
df_join = df_join.drop_duplicates(subset= keep_list, keep= "first")
```
